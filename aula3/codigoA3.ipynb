```python
import pandas as pd #Importa a biblioteca pandas
import numpy as np #Importa a biblioteca numpy
import matplotlib.pyplot as plt #Importa a biblioteca math.pyplot
```

```python
dados_x = np.linspace(0,10,10) #gere dados em x de 10 em 10 valores 
dados_y = dados_x*dados_x #Multiplica os dados de x e armazena em y
```

```python
plt.figure(figsize=(10,8), facecolor='gray') #Cria uma figura
Ax = plt.axes() #Cria os eixos para o gráfico
Ax.set_facecolor('white') #Cor do fundo do gráfico
plt.plot(dados_x,dados_y,color='red',marker = 'o', linestyle='dashed',markeredgecolor='black') #Plota o gráfico em x e y
```

```python
plt.rcParams.update({'font.size':15}) #função que permite ajustar
plt.figure(figsize=(12,12)) #Defini um tamanho para a figura
linhas = 2 #Cria 2 linhas
colunas = 2 #Cria 2 colunas
plt.subplot(linhas,colunas,1) #Subplot para plotar gráficos um ao lado do outro
plt.plot(dados_x,'r--', label='dados_x') #label rótulo para os dados
plt.legend(loc='best') #Cria uma legenda para o gráfico
plt.xlabel('Eixo x') #Eixo x
plt.xlabel('Eixo y') #Eixo y
plt.title('Gráfico 1') #Título para o gráfico

#Gráfico 2
plt.subplot(linhas,colunas,2) #Subplot para plotar gráficos um ao lado do outro
plt.legend(loc='best') #Cria uma legenda para o gráfico
plt.plot(dados_x,dados_y,'b-', label='dados_x \n e y') #label rótulo para os dados
plt.xlabel('Eixo x') #Eixo x
plt.xlabel('Eixo y') #Eixo y
plt.title('Gráfico 2')

#Gráfico 3
plt.subplot(linhas,colunas,3)
plt.plot(dados_x*dados_y,'g-',label='dados_x^2')
plt.legend(loc='lower right') #Cria uma legenda para o gráfico
plt.xlabel('Eixo x') #Eixo x
plt.xlabel('Eixo y') #Eixo y
plt.title('Gráfico 3')

#Gráfico 4
plt.subplot(linhas,colunas,4)
plt.plot(dados_y,'c--', label="dados_y")
plt.legend(loc='lower right') #Cria uma legenda para o gráfico
plt.xlabel('Eixo x') #Eixo x
plt.xlabel('Eixo y') #Eixo y
plt.title('Gráfico 4')

%config inlineBackend.figure_formats=['svg']
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(8,8))
linhas = 2
colunas = 2

plt.show()

```

```python
#Gráfico de barras
figura = plt.figure()#Cria uma variável chamada figura
grafico = figura.add_axes([0,0,0.7,0.7]) #Cria os eixos para o gráfico 
#Cria rotulos 
rotulos = ['A','B','C','D','E','F','G','H','I','J']   
#Plota o gráfico
grafico.bar(rotulos,dados_x,color='red') #Color altera a cor do gráfico
```

```python
grafico.bar(rotulos,dados_x,color='red') #Dados 1
grafico.bar(rotulos,dados_y*0.1,color='blue') #Dados 2
figura
```

```python
grafico.bar(dados_x,dados_y/2,color=['red','blue']) #Exibi os dados um do lado do outro
figura
```

```python
plt.boxplot(dados_x) #Plota o Boxplot   
plt.show() #Remove o 
```

```python
#Definindo nome do label x
plt.xlabel('Eixo x',#Nome do label
           fontweight='bold', #estilo
           fontsize='large', #tamanho da fonte
           fontfamily='fantasy', #tipo da fonte
           color = 'blue' #Cor da fonte
           )
#Definindo nome do label y
plt.ylabel('Eixo y',#Nome do label
           fontweight='bold', #estilo
           fontsize='large', #tamanho da fonte
           fontfamily='fantasy', #tipo da fonte
           color = 'blue' #Cor da fonte
           )
#Definindo o titulo
plt.title('Gráfico Simples',#titulo do gráfico  
           fontweight='bold', #estilo
           fontsize='xx-large', #tamanho da fonte
           fontfamily='fantasy', #tipo da fonte
           color = 'blue' #Cor da fonte
           )
plt.plot(dados_x,dados_y, #dados do gráfico
         color = 'red', linewidth='2' # Espessura da linha
         markersize=5, #Tamanho do marcador
         marker='o', #Tipo de marcador
         linestyle='dashed'
         )
```
