```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Importa da biblioteca sklearn o modelo de regressão linear
```

```python
#Teste do modelo criado
renda = np.array([1200,5000,8000,10000,30000]).reshape(-1, 1) 
gastos = np.array([750,1500,4000,6000,18000]) 
modelo = LinearRegression() #Armazena a regressão linear no modelo
modelo.fit(renda,gastos)#Treina o modelo de regressão linear
```

```python

renda_teste = np.array([1000,2000,2300,5000,16000]).reshape(-1,1)
#Cria modelo e treina o modelo de regressão linear

# Fazer previsões com base na vazão de água
gastos_previsto = modelo.predict(renda_teste)
```

```python
# Imprimir as previsões
for i in range(len(renda_teste)):
 print(f"Renda: {renda_teste[i][0]} - Gastos previsto: {gastos_previsto[i]} ")
plt.scatter(renda, gastos, color='blue', label='Dados de Treinamento')
plt.plot(renda_teste,gastos_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Relação entre Renda e Gastos')
plt.legend()
plt.grid(True)
plt.show()
```
