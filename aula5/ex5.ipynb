```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Importa da biblioteca sklearn o modelo de regressão linear
```

```python
#Teste do modelo criado
consumo = np.array([50,140,30,75,123]).reshape(-1, 1) 
demanda = np.array([100,200,80,100,160]) 
modelo = LinearRegression() #Armazena a regressão linear no modelo
modelo.fit(consumo,demanda)#Treina o modelo de regressão linear
```

```python

consumo_teste = np.array([10,30,50,70,100]).reshape(-1,1)
demanda_prevista = modelo.predict(consumo_teste)
```

```python
# Imprimir as previsões
for i in range(len(consumo_teste)):
 print(f"Consumo: {consumo_teste[i][0]} - Demanda prevista: {demanda_prevista[i]} ")
plt.scatter(consumo, demanda, color='blue', label='Dados de Treinamento')
plt.plot(consumo_teste,demanda_prevista, color = 'red',label='Dados de teste')
plt.xlabel('Consumo')
plt.ylabel('Demanda')
plt.title('Relação entre Consumo e Demanda')
plt.legend()
plt.grid(True)
plt.show()
```
