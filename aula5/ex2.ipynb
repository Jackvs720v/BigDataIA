```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Importa da biblioteca sklearn o modelo de regressão linear
```

```python
#Teste do modelo criado
quente = np.array([100,200,300,400,500,600,700]).reshape(-1, 1) # Quente 
temperatura = np.array([30,35,40,45,60,100,150]) #Temperatura do forno
modelo = LinearRegression() #Armazena a regressão linear no modelo
modelo.fit(quente,temperatura)#Treina o modelo de regressão linear
```

```python

calor_teste = np.array([10,30,50,70,80,100]).reshape(-1,1)
#Cria modelo e treina o modelo de regressão linear

# Fazer previsões com base na vazão de água
temperatura_prevista = modelo.predict(calor_teste)
```

```python
# Imprimir as previsões
for i in range(len(calor_teste)):
 print(f"Calor: {calor_teste[i][0]} m3/min - Temperatura prevista: {temperatura_prevista[i]}")
plt.scatter(quente, temperatura, color='blue', label='Dados de Treinamento')
plt.plot(calor_teste,temperatura_prevista, color = 'red',label='Dados de teste')
plt.xlabel('Temperatura')
plt.ylabel('Calor')
plt.title('Relação entre Calor e Temperatura')
plt.legend()
plt.grid(True)
plt.show()
```
