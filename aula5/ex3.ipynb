```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Importa da biblioteca sklearn o modelo de regressão linear
```

```python
#Teste do modelo criado
area = np.array([50,100,120,300]).reshape(-1, 1) # Quente 
valor = np.array([180.000,300.000,375.000,600.000]) #Temperatura do forno
modelo = LinearRegression() #Armazena a regressão linear no modelo
modelo.fit(area,valor)#Treina o modelo de regressão linear
```

```python

area_teste = np.array([10,30,40,100]).reshape(-1,1)
#Cria modelo e treina o modelo de regressão linear

# Fazer previsões com base na vazão de água
valor_previsto = modelo.predict(area_teste)
```

```python
# Imprimir as previsões
for i in range(len(area_teste)):
 print(f"Área: {area_teste[i][0]} M² - Valor previsto: {valor_previsto[i]} ")
plt.scatter(area, valor, color='blue', label='Dados de Treinamento')
plt.plot(area_teste,valor_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Área')
plt.ylabel('Valor')
plt.title('Relação entre Área e Valor do Imóvel')
plt.legend()
plt.grid(True)
plt.show()
```
