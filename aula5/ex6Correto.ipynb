```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dados de vendas
dic_vendas = {
    'Lojas': ['A', 'B', 'C', 'D'],
    
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abril': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 290, 120],
    'Jul': [150, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 200, 350, 190],
    'Dez': [300, 350, 400, 250],
    'Chave': ['AA', 'BB', 'CC', 'DD']
    
}
df_vendas = pd.DataFrame(dic_vendas)

# Definindo os meses para o eixo x
meses = ['Jan', 'Fev', 'Mar', 'Abril', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out',
          'Nov', 'Dez']
```

```python

# Configurando o grfico
plt.figure(figsize=(10, 6))

# Iterando por cada loja para plotar seu histrico de vendas
for i, loja in enumerate(df_vendas['Lojas']):
    plt.plot(meses, df_vendas.iloc[i, 1:-1], label=f'Loja {loja}', marker='o')

# Adicionando ttulos e rtulos
plt.title('Histrico de Vendas de Janeiro a Dezembro')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.legend(title="Lojas")
plt.grid(True)

# Exibindo o grfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
