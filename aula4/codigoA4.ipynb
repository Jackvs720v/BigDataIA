```python
import pandas as pd; #Importa a pandas
import matplotlib.pyplot as plt; #Importa a matplotlib
```

```python
dados = pd.read_csv('imigrantes_canada.csv'); #Importa a base de dados
```

```python
dados.tail() #Mostra as 4 últimas linhas
```

```python
dados.info()
```

```python
dados.set_index(['País'],inplace=True) #Seta a coluna Pais como index
anos = list(map(str,range(1980,2013))) #Transforma os anos em String utilizando função map
```

```python
print(anos) 
dados.tail(5)
```

```python
brasil = dados.loc['Brasil',anos] #Filtro que pega os dados somente do Brasil dados.loc['coluna',desejada]
```

```python
brasil 
```

```python
brasil_dic = {
    'ano': brasil.index.tolist(), #Pega os anos
    'imigrantes': brasil.values.tolist() #Pega a quantidade de imigrantes
}
```

```python
#Criando dataFrame para o Brasil
brasil_df = pd.DataFrame(brasil_dic)
```

```python

brasil_df.head()
```

```python
#Plotando o número de imigrantes do Brasil para o Canadá
plt.plot(brasil_df['ano'],brasil_df['imigrantes']) #plota o gráfico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração do Brasil para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
argentina = dados.loc['Argentina',anos]
```

```python
argentina_dic = {
    'ano': argentina.index.tolist(), #Pega os anos
    'imigrantes': argentina.values.tolist() #Pega a quantidade de imigrantes
}
```

```python
argentina_df = pd.DataFrame(argentina_dic)
```

```python
brasil_df.head()
```

```python
#Plotando o número de imigrantes do Brasil para o Canadá
plt.plot(argentina_df['ano'],argentina_df['imigrantes']) #plota o gráfico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14) # ajuste de valores para o eixo x
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores para o eixo y
plt.xlabel("Anos",fontsize=14) # titulo eixo x
plt.title('Imigração do Brasil para o Canadá',fontsize =18) # adiciona título no gráfico
plt.ylabel("Número de imigrantes",fontsize=14) # titulo eixo y
plt.show() #exibe a figura
```

```python
plt.figure(figsize=(10,5))
plt.plot(brasil_df['ano'],brasil_df['imigrantes'],color = 'red',label='Brasil')
plt.plot(argentina_df['ano'],argentina_df['imigrantes'],color = 'blue',label='Argentina')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel('Anos',fontsize=14)
plt.legend()
plt.savefig('Dados Imigração.jpg')
plt.ylabel('Número de imigrantes',fontsize=14)
plt.title('Imigração da Argentina e Brasil para o Canadá')
plt.show()

```

```python
plt.bar(brasil_df['ano'], brasil_df['imigrantes']) #plt.bar gráfico de barras 
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.title('Imigração do Brasil para o Canadá')
plt.show()
```

```python
df_comparacao = dados.loc[['Brasil','Argentina'],anos]  #filtro que pega duas colunas do dataset
```

```python
df_comparacao #Dataset de comparação
```

```python
df_comparacao = df_comparacao.T #Aplica a matriz transposta do dataframe
```

```python
df_comparacao['T'] = df_comparacao['Brasil'] + df_comparacao['Argentina'] #Soma o total de imigrantes
```

```python
plt.figure(figsize=(10,10))
plt.plot(df_comparacao['Brasil'], label='Brasil', color='green')
plt.plot(df_comparacao['Argentina'], label='Argentina', color='blue')
plt.plot(df_comparacao['T'], label='total', color='red')
plt.legend()
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14)
plt.title("Imigração do Brasil e \n Argentina para o Canadá",fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8)) #Cria a variavel ax para armazenar o subPlot
ax.plot(brasil_df['ano'],brasil_df['imigrantes'],lw=3,color='green')
#ax - Axes por conta que agora foi criada uma váriavel
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #Seta o eixo x ao espaçamento do gráfico
ax.xaxis.set_tick_params(labelsize=12) #Altera o tamanho da fonte do gráfico
ax.yaxis.set_tick_params(labelsize=12) #Altera o tamanho da fonte do eixo y
ax.set_xlabel('Anos',fontsize=12)
ax.set_ylabel('Número de imigrantes',fontsize=12)
ax.set_title("Imagrantes do Brasil para o Canadá", fontsize=18)
#ax.legend(loc='upper right')
#ax.legend(loc='upper left')
#ax.legend(loc='lower right')
#ax.legend(loc='lower left')
ax.legend(loc='best')
ax.grid() #Add linhas no gráfico que ajudam a interpretar



```

```python
#Cria um subplot com duas figuras no gráfico
fig, axs = plt.subplots(1,2,figsize=(15,5)) #Cria uma várivel axs que corresponde a eixos diferentes no gráfico subplot (quantidade de linhas, colunas, figsize - Tamanho da Figura)
axs[0].plot(brasil_df['ano'],brasil_df['imigrantes']) #Plota o primeiro gráfico na posição zero
axs[0].set_title('Imigrantes do Brasil para o Canadá de \n 1980 - 2013',fontsize=18)
axs[0].set_xlabel('anos',fontsize=12)
axs[0].set_xlabel('Número de imigrantes',fontsize=12)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))#Cria o intervalo de 5 em 5 no eixo x
axs[0].grid() #Cria o grid do gráfico

#Gráfico 2
axs[1].boxplot(brasil_df['imigrantes'])
axs[1].set_title('Box Plot de imigração do Brasil \n para o Canadá',fontsize=18)
axs[1].set_xlabel('Brasil', fontsize=14)
axs[1].set_ylabel('Nº do imigrantes',fontsize=12)
axs[1].grid()

plt.show()
```
